```python
df.dropna(how = "any", inplace = True) #Removing the missing values because the missing values arent significant for the analysis
```

```python
df.isnull().sum() #checking the missing values again
```

```python
df.describe(include = "all") #Summary Statistics
```

```python
df.columns = df.columns.str.lower() #Making the columns lowercase
```

```python
df.columns = df.columns.str.replace(" ","_") #Replacing spaces with _ in column names
```

```python
df.drop("brand", axis =1, inplace = True) #Dropping the brand column because it contains only the company's name
```

```python
df.drop("url", axis =1, inplace = True) #Dropping the url column because it isn't necessary
```

```python
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float) #Changing the data type of price into float
```

```python
df.set_index('product_id', inplace=True) #Setting the product_id as index
```

```python
#Feature Engineering
```

```python
df['price_elasticity'] = df['sales_volume'] / df['price'] #Creating a price_elasiticity column
```

```python
df['price_category'] = df['price'].apply(lambda p: 'Low' if p < 35 else ('Medium' if p < 70 else 'High')) #Creating a price_category column
```
